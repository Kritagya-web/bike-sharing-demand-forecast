```python
from scipy.stats.contingency import association

categorical_cols = ['season', 'yr', 'mnth', 'hr',
    'holiday', 'weekday', 'workingday', 'weathersit']

cat_association = pd.DataFrame(index = categorical_cols, columns = categorical_cols)

# Compare each pair of categorical features with others using Cramér's V

for col1 in categorical_cols:
    for col2 in categorical_cols:
        if col1 == col2:
            cat_association.loc[col1, col2] = 1
        else:
            table = pd.crosstab(hour_df[col1], hour_df[col2])
            score = association(table, method = 'cramer', correction = False)
            cat_association.loc[col1, col2] = score

cat_association = cat_association.astype(float)
plt.figure(figsize=(10,6))
sns.heatmap(cat_association, annot=True, cmap = 'Blues', fmt = '.2f')
plt.title("Categorical Feature Association Heatmap (Cramer's V)")
plt.show()
```

### STEP 3 — Data Cleaning + Preprocessing

```python
# Remove useless features like instant, casual, registered, atemp
# Define target and features
# Handling Categorical Features
# Handling time features
# Scale numerical features
# Train Test Split
```

```python
df = hour_df.copy()
```

```python
df = df.drop(columns=['instant', 'casual', 'registered', 'atemp', 'dteday'])
```

```python
y = df['cnt']
X = df.drop(columns=['cnt'])
```

```python
X.head()
```

```python
# Handling Cyclical Features
# Imp as Hr 23 and 0 are pretty close to each other 
# Numerically same but different as 0 ad 23 are far 
X['hr_sin'] = np.sin(2*np.pi*X['hr']/24)
X['hr_cos'] = np.cos(2*np.pi*X['hr']/24)

# For Month as 12 and 1 are close to each other but numerically they are far
X['mnth_sin'] = np.sin(2*np.pi*X['mnth']/12)
X['mnth_cos'] = np.cos(2*np.pi*X['mnth']/12)

# For weekday as 0 and 6 are close to each other but numerically they are far
X['weekday_sin'] = np.sin(2*np.pi*X['weekday']/7)
X['weekday_cos'] = np.cos(2*np.pi*X['weekday']/7)
```

```python
X.head()
```

```python
X = X.drop(columns = ['hr', 'mnth', 'weekday'])
```

```python
X.head()
```

```python
# Handling Categorical Features - To Avoid Multicollinearity and Dummy Variable Trap
categorical_cols = ['season', 'yr', 'holiday', 'workingday', 'weathersit']

X = pd.get_dummies(X, columns = categorical_cols, drop_first=True)
```

```python
# Scaling the numerical features
from sklearn.preprocessing import StandardScaler

numeric_cols = ['temp', 'hum', 'windspeed']

scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```

```python
print(X.shape, y.shape)
```

```python
# Time based Train Test Split - As it is a time series kind of data
# 80 percent data for training and 20 percent for testing

split_index = int(0.8*len(X))
X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]
```

```python
print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
# # Model A - Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
```

```python
# Evaluate the model based of MAE, RMSE and R2 Score

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)
```

```python
# # Compare actual vs predicted values

plt.figure(figsize=(10,6))
plt.scatter(y_test, y_pred, alpha = 0.3)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
# # Residual Analysis

residual = y_test - y_pred 
plt.Figure(figsize=(8,6))
plt.scatter(y_pred, residual, alpha=0.3)
plt.axhline(0, color = 'red')
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
# 5 - Feature Engineering (Improving the Model Input)
# - We would try to add non-linear relationships

X.columns
```

```python
# A great feature to have would be to create a new feature for hr X workingdays as 
X['hr_workingday'] = X['hr_sin'] * X['workingday_1']
```

```python
# Another good feature to have would be weather X temp or hr X weather or workingday X weather because it assumes a cnt ∝ temp

# Too cold -> Low demand
# Pleasant -> High Demand
# Too hot -> Lower Demand
```

```python
# Good weather and peak hr would get highlighted in this.
X['hr_weather_clear'] = X['hr_sin'] * X['weathersit_2']
# Weather and Temperature go good together
X['temp_weather'] = X['temp'] * X.filter(like ='weathersit_').sum(axis = 1)
X['workingday_weather'] = X['workingday_1'] * X.filter(like ='weathersit_').sum(axis = 1)
```

```python
X['temp_squared'] = X['temp'] ** 2
X['hum_squared'] = X['hum'] ** 2
```

```python
# Step 1: Do ALL feature engineering on X

# Step 2: THEN split

split_index = int(0.8*len(X))

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]

y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]
```

```python
# Model A - Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
```

```python
# Evaluate the model based of MAE, RMSE and R2 Score

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R2 Score:", r2)

# Previous One
# Mean Absolute Error: 120.9223916412678
# Root Mean Squared Error: 164.491709203897
# R2 Score: 0.44341608759971496
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
rf_model = RandomForestRegressor(n_estimators=200, max_depth = None, min_samples_split = 2, min_samples_leaf = 1, random_state = 42, n_jobs = -1)
```

```python
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)

print("Random Forest MAE:", rf_mae)
print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2:", rf_r2)
```

```python
# Plot Actual Vs Predicted
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(y_test, rf_pred, alpha = 0.3)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Random Forest : Actual Vs Predicted")
plt.show()
```

```python
# Improvement Option 1 - Tune the Random Forest Model

rf_model_tuned = RandomForestRegressor(n_estimators=300, max_depth=15, min_samples_split=10, min_samples_leaf=5, max_features='sqrt', random_state=42, n_jobs=-1)
```

```python
rf_model_tuned.fit(X_train, y_train)

rf_tuned_pred = rf_model_tuned.predict(X_test)

rf_tuned_r2 = r2_score(y_test, rf_tuned_pred)
rf_tuned_mae = mean_absolute_error(y_test, rf_tuned_pred)
rf_tuned_rmse = np.sqrt(mean_squared_error(y_test, rf_tuned_pred))

print("Tuned RF R2:", rf_tuned_r2)
```

```python
# Expected outcome
# Train R² ↓ (good)
# Test R² ≈ same or slightly ↑ (great)
# Overfitting gap ↓

# It became underfitting now as it reduced the gap by a lot of margin 
# max_depth=15 → too shallow
# min_samples_leaf=5 → too smooth
# min_samples_split=10 → too restrictive

# 👉 Model became too simple
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(n_estimators = 700, learning_rate = 0.03, max_depth = 6, subsample = 0.8, colsample_bytree = 0.8, random_state = 42)
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
```

```python
xgb_mae = mean_absolute_error(y_test, xgb_pred)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
xgb_r2 = r2_score(y_test, xgb_pred)

print("XGBoost MAE:", xgb_mae)
print("XGBoost RMSE:", xgb_rmse)
print("XGBoost R2:", xgb_r2)
```

```python
# Random Forest improved performance significantly by capturing non-linear relationships, but exhibited overfitting due to deep trees. When over-regularized, it underfit. XGBoost provided the best balance between bias and variance by learning sequentially and applying regularization, resulting in the highest generalization performance.
```
